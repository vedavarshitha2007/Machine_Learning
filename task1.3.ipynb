```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df=sns.load_dataset("iris")

print(df.head())
```

```python
#Histogram
plt.figure(figsize=(6,4))
sns.histplot(df["sepal_length"],bins=15,kde=True)
plt.title("Histogram of Sepal Length(24eu01131)")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="species",data=df)
plt.title("Species Count(24eu01131)")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(y="petal_length", data=df)
plt.title("Box Plot of Petal Length(24eu01131)")
plt.show()

plt.figure(figsize=(6,4))
sns.violinplot(y="petal_width", data=df)
plt.title("Violin Plot of Petal Width(24eu01131)")
plt.show()

plt.figure(figsize=(6,4))
sns.kdeplot(df["sepal_width"], fill=True)
plt.title("Density Plot of Sepal Width(24eu01131)")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x="sepal_length",y="petal_length",hue="species", data=df)
plt.title("Sepal Length vs Petal Length(24eu01131)")
plt.show()
```

```python
sns.pairplot(df,hue="species")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.heatmap(df.corr(numeric_only=True),
            annot=True,
            cmap="coolwarm")
plt.title("Correlation Heatmap(24eu01131)")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x="species", y="petal_length", data=df)
plt.title("Petal Length by Species(24eu01131)")
plt.show()


plt.figure(figsize=(6,4))
sns.violinplot(x="species", y="sepal_width", data=df)
plt.title("Sepal Width by Species(24eu01131)")
plt.show()


plt.figure(figsize=(6,4))
sns.barplot(x="species", y="petal_width", data=df)
plt.title("Average Petal Width by Species(24eu01131)")
plt.show()
```
